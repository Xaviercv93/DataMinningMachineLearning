### Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column.

```r
# Set seed for reproducibility
set.seed(123)

# Create a 2-dimensional array with random integer values
array <- matrix(sample(1:10, 3 * 4, replace = TRUE), nrow = 3, ncol = 4)

print("Imprime la matriz")
print(array)
cat("\n")
print("Calcular la suma de cada fila")
row_sums <- rowSums(array)
print(row_sums)
cat("\n")
print("Calcular la suma de cada columna")
col_sums <- colSums(array)
print(col_sums)
```

### Create a DataFrame with 3 columns: "Name", "Age", and "City". Add at least 5 rows of data to the DataFrame. Then, filter the DataFrame to only include rows where the person's age is greater than or equal to 30.

```r
# Crear el DataFrame
df <- data.frame(
  Name = c("John", "Alice", "Michael", "Emma", "Daniel"),
  Age = c(25, 32, 40, 28, 35),
  City = c("New York", "London", "Paris", "Sydney", "Tokyo")
)
print("Imprimir el DataFrame")
print(df)
cat("\n")
# filtrar un DataFrame en R, se utiliza la notación de subconjunto para seleccionar las filas donde el valor de la columna "Age" es mayor o igual a 30.
filtered_df <- df[df$Age>=30,]
print("Imprimir el DataFrame filtrado por edad >= 30 en forma 1")
print(filtered_df)
cat("\n")
#la función subset() es otra forma común de filtrar un DataFrame en R
filtered_df2 <- subset(df, Age >= 30)
print("Imprimir el DataFrame filtrado por edad >= 30 en forma 2")
print(filtered_df2)


```

### Load a CSV file into a DataFrame and then calculate the mean, median, and mode of one of the columns in the DataFrame.

```r
print("Imprimir el dataframe calcular que contiene los valores de un archivo CSV")
calcular <- read.csv('/kaggle/input/afeitadoras-hombres-en-amazon/shavers_for_men_amazon.csv', nrows = 20, header=TRUE,sep=",", dec = ",")
calcular
cat("\n")
# Convertir la columna de carácter a numérica
#calcular$price <- as.numeric(calcular$price)
# Verificar que la conversión se haya realizado correctamente
#calcular
# valor es un vector que contiene los datos de score
valor <- calcular$score
mean_value <- mean(valor, na.rm = TRUE)  # Calculate the mean
print("Imprimir el valor promedio de la columna score")
print(mean_value)  # Print the mean
cat("\n")
median_value <- median(valor, na.rm = TRUE)  # Calculate the median
print("Imprimir el valor de la media de la columna score")
print(median_value)  # Print the median
cat("\n")
# mode_value <- calcular$score %>% table() %>% as.data.frame() %>% filter(Freq == max(Freq)) %>% pull(score)  # Calculate the mode
score_counts <- table(calcular$score)  # Tabla de frecuencias
max_freq <- max(score_counts)  # Frecuencia máxima
mode_values <- as.numeric(names(score_counts[score_counts == max_freq]))  # Valor(es) de moda
print("Imprimir el valor de la moda de la columna score")
print(mode_values)  # Print the mode
```

### Create a scatter plot of random x and y values between 1 and 100.

```r
# Generar valores aleatorios para los ejes x e y
# la función runif permite generar 100 valores aleatorios entre 1 y 100
x <- runif(100, min = 1, max = 100)
y <- runif(100, min = 1, max = 100)

# Crear el gráfico de dispersión
# pch es un número entero o un carácter que representa el símbolo utilizado para representar los puntos en el gráfico
# col es un parámetro que indica el color de los puntos en el gráfico
plot(x, y, main = "Scatter Plot con valores aleatorios", xlab = "eje X", ylab = "eje Y", pch = "@", col = "blue")
```
