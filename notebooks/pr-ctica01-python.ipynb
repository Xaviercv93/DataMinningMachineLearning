## Create an array of 10 random integers between 1 and 100, and then calculate the mean and standard deviation of the array

```python
arreglo_random = np.random.randint(1, 101, size=10)
print("Arreglo Random:", arreglo_random, type(arreglo_random))
media= mean = np.mean(arreglo_random)
print("Media:", media)
desviacion_estandar= np.std(arreglo_random)
print("Desviacion Estandar:", desviacion_estandar)



```

## Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column

```python
arreglo_random = np.random.randint(1, 11, size=(3, 4))

suma_filas = np.sum(arreglo_random, axis=1)

suma_columnas = np.sum(arreglo_random, axis=0)

print("Arreglo Random:")
print(arreglo_random)
print("\nSuma filas:")
print(suma_filas)
print("\nSuma Columnas:")
print(suma_columnas)
```

## Create a DataFrame with 3 columns: "Name", "Age", and "City". Add at least 5 rows of data to the DataFrame. Then, filter the DataFrame to only include rows where the person's age is greater than or equal to 30.

## Load a CSV file into a DataFrame and then calculate the mean, median, and mode of one of the columns in the DataFrame

```python
df = pd.read_csv('/kaggle/input/servicearea/servicearea.csv',sep=',')
print(df.head())

# Calcular la media de una columna
mean_value = df['VersionNum'].mean()

# Calcular la mediana de una columna
median_value = df['VersionNum'].median()

# Calcular la moda de una columna
mode_value = df['VersionNum'].mode()

# Imprimir los resultados
print('Media:', mean_value)
print('Mediana:', median_value)
print('Moda:', mode_value)
```

##  Create a scatter plot of random x and y values between 1 and 100

```python
# Generar valores aleatorios para x e y
x = [random.randint(1, 100) for _ in range(100)]
y = [random.randint(1, 100) for _ in range(100)]

# Crear el gráfico de dispersión
plt.scatter(x, y)

# Agregar etiquetas y título al gráfico
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de dispersión')

# Mostrar el gráfico
plt.show()
```
