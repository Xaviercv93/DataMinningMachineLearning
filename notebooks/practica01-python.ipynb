## (r) Select the days in May with a temperature above 70 Fahrenheit.

```python
# Convertir la columna "Month" en formato de categoría
airquality['Month'] = airquality['Month'].astype('category')

# Filtrar los días de mayo con temperatura superior a 70 grados Fahrenheit
may_high_temp = airquality[(airquality['Month'] == 5) & (airquality['Temp'] > 70)]

# Imprimir los resultados
print(may_high_temp)
```

## (s) Create a new attribute TempC which represents the temperature values in Celsius.

```python
# Convertir la temperatura de Fahrenheit a Celsius
airquality['TempC'] = (airquality['Temp'] - 32) * (5/9)

# Imprimir los resultados
print(airquality)
```

## (t) Inspect which were the 30 hottest days.

```python
# Ordenar el conjunto de datos por la columna "TempC" en orden descendente
hottest_days = airquality.sort_values(by='TempC', ascending=False).head(30)

# Imprimir los resultados
print(hottest_days)
```

## (u) Inspect which were the hottest days, but also with the highest ozone values.


```python
# Ordenar el conjunto de datos por las columnas "TempC" y "Ozone" en orden descendente
hottest_ozone_days = airquality.sort_values(by=['TempC', 'Ozone'], ascending=[False, False])

# Imprimir los resultados
print(hottest_ozone_days)
```

## (v) Inspect the number of days for which there was a register for each month

```python
# Contar el número de días para cada mes
days_per_month = airquality['Month'].value_counts()

# Imprimir los resultados
print(days_per_month)
```

## (w) For each month, obtain the minimum and the maximum temperature registered in Celsius.

```python
# Agrupar por mes y calcular la temperatura mínima y máxima
temp_stats = airquality.groupby('Month')['TempC'].agg(['min', 'max'])

# Imprimir los resultados
print(temp_stats)
```

## (x) Obtain the average of the following parameters by month: temperature in celsius, wind, solar radiation and ozone.

```python
# Agrupar por mes y calcular el promedio de los parámetros
avg_params = airquality.groupby('Month')[['TempC', 'Wind', 'Solar.R', 'Ozone']].mean()

# Imprimir los resultados
print(avg_params)
```

## (y) What values did you obtain regarding ozone and solar radiation attributes? Why? Make the necessary change so that you get the average of the registered values

```python
# Calcular el promedio de todos los valores registrados para los atributos de ozono y radiación solar
avg_ozone = airquality['Ozone'].mean()
avg_solar_radiation = airquality['Solar.R'].mean()

# Imprimir los resultados
print("Promedio de ozono:", avg_ozone)
print("Promedio de radiación solar:", avg_solar_radiation)
```
